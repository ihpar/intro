```python
# kullanıcıdan, kullanıcı "q" yazana kadar sayılar alıp toplam sayıyı hesaplayalım
# 3, 5, 1, q -> 9
# 1, 7, q -> 8
# 2, 4, 6, 8, 10, q -> 30

toplam = 0
while True:
    girdi = input("Bir sayı veya q giriniz:")
    if girdi == "q":
        break
    toplam += int(girdi)

print(f"Girdiğiniz bütün sayıların toplamı = {toplam}") # python f-strings
```

```python
# kullanıcıdan, kullanıcı "q" yazana kadar sayılar alıp sayıların arit. ort. hesaplayalım
# 3, 5, 1, q -> 3
# 1, 7, q -> 4
# 2, 4, 6, 8, 10, q -> 6

toplam = 0
sayi_adedi = 0

while True:
    girdi = input("Bir sayı veya q giriniz:")
    if girdi == "q":
        break

    toplam += int(girdi)
    sayi_adedi += 1

ortalama = toplam / sayi_adedi
print("Girdiğiniz sayıların ortalaması =", ortalama)
```

```python
# listedeki en küçük sayıyı bulma

sayilar = [3, 1, 4, 2, -3, 5, 8, 4]
# print(min(sayilar))
# print(max(sayilar))

en_kucuk = sayilar[0]
en_buyuk = sayilar[0]
i = 1 
while i < len(sayilar):
    if sayilar[i] < en_kucuk:
        en_kucuk = sayilar[i]
    
    if sayilar[i] > en_buyuk:
        en_buyuk = sayilar[i]

    i += 1

print(en_kucuk, en_buyuk)
```
