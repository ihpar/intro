```python
""" 
Bilgisayar 1'den 10'a kadar bi sayı belirler.
3 seferde bu sayıyı bulmaya çalıştığımız bir oyun.

1)
hedef = 7
tahmin = 2 -> hayır, daha büyük

tahmin = 9 -> hayır, daha küçük

tahmin = 8 -> kaybettin, doğru sayı 7 idi.

1)
hedef = 5
tahmin = 1 -> hayır, daha büyük

tahmin = 7 -> hayır, daha küçük

tahmin = 5 -> kazandın tebrikler.

1, 2, 3, 4, 5, 6, 7, 8, 9
"""
import random
```

```python
hedef = random.randint(1, 9)
print(hedef)

tahmin = int(input("1 - 9 arasında bir sayi girin:"))

if tahmin == hedef:  # ilk deneme
    print("Tebrikler oyunu kazandın!")
else:
    if hedef > tahmin:
        print("Hayır, daha büyük bir sayı seçmelisin.")
        tahmin = int(input("1 - 9 arasında bir sayi girin:"))
        if tahmin == hedef:  # ikinci deneme
            print("Tebrikler oyunu kazandın!")
        else:
            # akış buraya ulaştıysa hedef tahminden büyüktür.
            if hedef > tahmin:
                print("Hayır, daha büyük bir sayı seçmelisin.")
                tahmin = int(input("1 - 9 arasında bir sayi girin:"))
                if tahmin == hedef:  # son deneme
                    print("Tebrikler, oyunu kazandın!")
                else:
                    print("Kaybettin :-(")
            else:
                print("Hayır, daha küçük bir sayı seçmelisin.")
                tahmin = int(input("1 - 9 arasında bir sayi girin:"))
                if tahmin == hedef:  # son deneme
                    print("Tebrikler, oyunu kazandın!")
                else:
                    print("Kaybettin :-(")
    else:
        print("Hayır, daha küçük bir sayı seçmelisin.")
```

```python
hedef = random.randint(1, 9)
print(hedef)

tahmin = int(input("1 - 9 arasında bir sayi girin:"))

if tahmin == hedef:  # ilk deneme
    print("Tebrikler oyunu kazandın!")
else:
    if hedef > tahmin:
        print("Hayır, daha büyük bir sayı seçmelisin.")
        tahmin = int(input("1 - 9 arasında bir sayi girin:"))
        if tahmin == hedef:  # ikinci deneme
            print("Tebrikler oyunu kazandın!")
        else:
            # akış buraya ulaştıysa hedef tahminden büyüktür.
            if hedef > tahmin:
                print("Hayır, daha büyük bir sayı seçmelisin.")
            else:
                print("Hayır, daha küçük bir sayı seçmelisin.")

            tahmin = int(input("1 - 9 arasında bir sayi girin:"))
            if tahmin == hedef:  # son deneme
                print("Tebrikler, oyunu kazandın!")
            else:
                print("Kaybettin :-(")

    else:
        print("Hayır, daha küçük bir sayı seçmelisin.")
```

```python
hedef = random.randint(1, 9)
tahmin = int(input("İlk tahmin:"))

bitti_mi = False

# ilk tur
if tahmin == hedef:
    print("İlk denemede bildin: 20 puan")
    bitti_mi = True
elif tahmin < hedef:
    print("Yukarı")
else:
    print("Aşağı")

# ikinci tur
if not bitti_mi:
    tahmin = int(input("İkinci tahmin:"))
    if tahmin == hedef:
        print("İkinci denemede bildin: 10 puan")
        bitti_mi = True
    elif tahmin < hedef:
        print("Yukarı")
    else:
        print("Aşağı")

# üçüncü tur
if not bitti_mi:
    tahmin = int(input("Son tahmin:"))
    if tahmin == hedef:
        print("Son denemede bildin: 5 puan")
    else:
        print("Maalesef kaybettin...")
```
