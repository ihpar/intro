```python
# 0<= sınav notu < 50 -> kötü; 50 <= sn < 75 -> orta; 75 <= sn -> iyi

notun = int(input("Notun ne?"))

if (notun < 0) or (notun > 100):
    print("Sınav notu 0 - 100 aralığında olmalıdır.")
elif (notun >= 0) and (notun < 50):
    print("Kötü almışsın")
elif (notun >=50) and (notun < 75):
    print("Ortalama almışsın")
else:
    print("İyi almışsın")

print("son")
```

```python
sayi = int(input("Bir sayı giriniz"))

if sayi % 2 == 0:
    print(sayi, "sayısı bir çift sayıdır.")
elif sayi % 3 == 0:
    print(sayi, "sayısı 3'ün tam katıdır.")
else:
    print(sayi, "sayisi ne çifttir ne de 3'ün tam katıdır.")
```
