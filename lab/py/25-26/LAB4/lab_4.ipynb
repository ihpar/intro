```python
i = 0
satir_sayisi = 12
sutun_saiyisi = 17
while i < satir_sayisi:
    if i == 0 or i == satir_sayisi - 1:
        j = 0
        while j < sutun_saiyisi:
            print("*", end=" ")
            j += 1
        print("")
    else:
        j = 0
        while j < sutun_saiyisi:
            if j == 0 or j == sutun_saiyisi - 1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
            j += 1
        print("")
    i += 1
```

```python
i = 0
satir_sayisi = 5
sutun_saiyisi = 7
while i < satir_sayisi:
    j = 0
    if i == 0 or i == satir_sayisi - 1:
        while j < sutun_saiyisi:
            print("*", end=" ")
            j += 1
    else:
        while j < sutun_saiyisi:
            if j == 0 or j == sutun_saiyisi - 1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
            j += 1
    print("")
    i += 1
```

```python
i = 0

while i < 3:
    j = 0
    while j < 2:
        k = 0
        while k < 4:
            print(f"i: {i}, j: {j}, k: {k}")  # f-strings
            k += 1
        j += 1
    i += 1
```
